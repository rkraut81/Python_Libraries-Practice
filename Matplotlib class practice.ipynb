```python
# Import relevant modules
import pandas as pd
import matplotlib.pyplot as plt
# Import dataset
data2 = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\datasets-master\datasets-master\Carseats.csv')
sales = data2['Sales']
age = data2['Age']
income = data2['Income']
price = data2['Price']

# Plot the data series

plt.scatter(sales, age,c='r',label='Age')

plt.scatter(sales, income,c='g', label='Income')

plt.scatter(sales, price,c='y', label='Price')


plt.legend()
plt.xlabel('Sales',size=12,fontweight='semibold')
plt.ylabel('Features',size=12,fontweight='semibold')
plt.title('CAR SEAT SALES', size=14, fontweight='semibold')
plt.xlim(0,20)
plt.ylim(0,200)
plt.savefig('Carsales.png')
# Show the figure
plt.show()
```

```python
#Adding trendline over scatterplot

x = list(range(10))
y = x + np.random.rand(10)-0.5

# Calculate the slope and y-intercept of the trendline
fit = np.polyfit(x,y,1)

# Add the trendline
yfit = [n*fit[0] for n in x] + fit[1]
plt.scatter(x,y)
plt.plot(yfit,'black')

plt.show()
```
