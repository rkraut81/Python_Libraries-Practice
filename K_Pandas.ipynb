# ACCESSING THE ELEMENTS

1. .loc - focus only on row indexes
2. .iloc - focus on row and column indexes
3. Through column names



```python
# loc returns results based on row indexes 

df.loc['R1']
```

```python
type(df.loc['R1'])
```

```python
#Data frames is the combination of many many columns and many rows. Atleast it should be greater than 1 column and 1 row.
# Data Series can either be one row  or one column
```

```python
#iloc returns results based on row indexes and column indexes
# all rows and all columns just like numpy
df.iloc[ : , : ]
```

```python
# to return only 3 rows and 2 columns

df.iloc[ 0:3 , 0:2 ]
```

```python
# As the above result is 3 rows and 2 columns, hence it is a dataframe . 
type(df.iloc[ 0:3 , 0:2 ])
```

```python
df.iloc[ 0:1 , 0:1 ]
```

```python
type(df.iloc[ 0:1 , 0:1 ])
```

### NOTE : iloc returns dataframe though it is one row and one column but loc returns DATA series 

```python
df.iloc[0:2  ,0 ]
```

```python
type(df.iloc[0:2  ,0 ])
```

# VARIOUS PARAMETERS UNDER READ_CSV

### 1. sep(' ; ') - use the sep keyword to put any character as separator as the csv file is not restrocted to comma(,)
### 2. usecols - use the specific column names you want to import in your data drame
### 3. dtype - to particularly assign types to the columns ( assign a dictionary of key value pairs)
### 4. index_col= <index column  : Index columns and training delimiters
### 5. escapechar= ' <char like //>'  

```python
# sep(' ; ') - use the sep keyword to put any character as separator as the csv file is not restrocted to comma(,)

test_df=pd.read_csv('ex1.csv',sep=';')
```

```python
## Index columns and training delimiters

data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
df=pd.read_csv(StringIO(data))
```

```python
df
```

```python
# To make the 4, 8 as index column we have to specify "index_col = 0"

df=pd.read_csv(StringIO(data),index_col=0)
```

```python
df
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
## Combining usecols and index_col
data = ('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow,')
```

```python
df=pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```

```python
df
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python

pd.read_csv(StringIO(data),escapechar='\\')
```

# Pickling¶
All pandas objects are equipped with to_pickle methods which use Python’s cPickle module to save data structures to disk using the pickle format.

If having huge dataset and lot of preprocessing is required , then we convert that datset into pickle.
Suppose if your kernel is restarted , u hv to execute all the code from beginning hence its better to do pickle of the datset to create datastructures within

```python

df_excel.to_pickle('df_excel')
```
