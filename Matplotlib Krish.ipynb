```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
# SCATTER PLOT

plt.scatter(x,y,c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D GRAPH')
plt.savefig('2d.png')
```

```python
# LINE PLOT 

y= x*x

plt.plot(x,y,'r*--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2d Diagram')
```

```python
# Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

```python
x=np.arange(1,11)
y=2*x+5
plt.plot(x,y,'r*--')
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#sine wave 

x=np.arange(0,4*np.pi,0.1 )
y=np.sin(x)

plt.title('Sine wave form')
plt.plot(x,y,'r--')
```

```python
#sine wave and cos wave in subplots

x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('Sine wave')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('Cosine wave')

plt.show()
plt.savefig('sincos.png')
```

```python
#BAR PLOTS

x = [2,8,15] 
y = [11,16,9]  

x2 = [3,9,16] 
y2 = [6,15,7] 

plt.bar(x,y,color = 'r')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
## HISTOGRAMS

a = np.array([22,87,5,45,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a, color='y', bins=20) 
plt.title("histogram") 
plt.show()
```

```python
## BOX PLOT 

data=[np.random.normal(0,std, 100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

```python
## Pie chart 

labels = 'python', 'r','java', 'c++'
sizes=[215,180,210,100]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode=(0.2,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
