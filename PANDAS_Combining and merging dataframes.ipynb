```python
import pandas as pd
import numpy as np
```

```python
#Userinfo
usercuisine=pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\usercuisine.csv')
payment= pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\userpayment.csv')
profile=pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\userprofile.csv')

# restaurant info
rest_accepts = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\chefmozaccepts.csv')
rest_cuisine = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\chefmozcuisine.csv')
rest_hours   = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\chefmozhours4.csv')
rest_parking = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\chefmozparking.csv')

# Loc and rating info
geo_loc = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\geoplaces2.csv')
rating_final = pd.read_csv(r'C:\Users\Aarav\Desktop\end to end data sets\restaurant-data-with-consumer-ratings\rating_final.csv')

```

```python
# full outer join

rest_full_outer_join=pd.merge(rest_accepts,rest_cuisine, on='placeID', how='outer')
```

```python
rest_full_outer_join.head()
```

```python
rest_inner_join = pd.merge(rest_accepts, rest_cuisine, on='placeID', how='inner')
```

```python
rest_inner_join.head()
```

```python
rest_right_join=pd.merge(rest_accepts,rest_cuisine,on = 'placeID', how='right')
```

```python
rest_right_join.head()
```

```python
rest_left_join=pd.merge(rest_accepts,rest_cuisine,on = 'placeID', how='left')
```
